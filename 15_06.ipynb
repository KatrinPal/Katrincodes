# Readlines


```python
filename = '/content/births.txt' #location

with open(filename, 'r') as file: # 'r' is for 'read'
  content = file.readlines() # here is readlines instead of read

  print('Standard File I/O') # I/O meaning input/output
  print(content)
```

```python
for line in content:
  print(line, end='')
```

```python
filename = input("Enter the filename with path: ")

try:
    with open(filename, 'r') as file:
        content = file.readlines()

    print('This is Standard File I/O')
    print(content)
except FileNotFoundError:
    print(f"Error: The file '{filename}' was not found.")
```

```python
with open('/content/births.txt', 'r') as file:
    births = file.readlines() #list, separated by a comma

with open('/content/deaths.txt', 'r') as file:
    deaths = file.readlines()
```

```python
len(births) == len(deaths)
```

```python
# accessing the number: int(birthline[4:])

for i in range(len(births)):
  birthline = births[i]
  deathline = deaths[i]

  birthnumber = int(birthline[4:])
  deathnumber = int(deathline[4:])

  growth = birthnumber- deathnumber
  month = birthline[3:]
  if growth > 0:
    print(month, growth)
```
